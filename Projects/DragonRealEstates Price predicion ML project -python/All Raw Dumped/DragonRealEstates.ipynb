# Feature Scaling

primarily, we have two types of feature scaling methods :-

    1. min-max value (Normalization)
        (value-min)/(max-min)
        for tis we have a class MinMaxScaler from sklearn
        
    2. Standardization 
        (value-mean)/standardDivision
        for tis we have a class StandardScaler from sklearn
        this is more reliable coz in MinMaxScaler, if one value goes wrong then the whole dataset is affected but here in 
        StandardScaler it doesnt affect other valuse

# Pipeline

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

myPipe = Pipeline([
    ('imputer',SimpleImputer(strategy='median')),
    ('std',StandardScaler()),
])
```

```python
hr = myPipe.fit_transform(hrTemp)
```

```python
hr
```

# Selecting A Desired Model For Dragon Real Estates

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model = RandomForestRegressor()
```

```python
model.fit(hr,housing_labels)
```

# Evaluating The Model

```python
from sklearn.metrics import mean_squared_error
```

```python
predictions = model.predict(hr)
```

```python
MSE = mean_squared_error(housing_labels,predictions)
rmse = np.sqrt(MSE)
```

```python
MSE
```

# Cross Validation - Better Evaluation Technique

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(model,hr,housing_labels, scoring='neg_mean_squared_error',cv=10)
rmseScores = np.sqrt(-scores)
```

```python
def eval (scores):
    print('-------------------------------')
    print('Scores : ',scores)
    print('-------------------------------')
    print('Mean : ', scores.mean())
    print('-------------------------------')
    print('STD : ', scores.std())
    print('-------------------------------')
    print('RMSE : ',rmseScores)
    print('-------------------------------')
```

```python
eval(rmseScores)
```
